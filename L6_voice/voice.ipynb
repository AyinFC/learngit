```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('./voice.csv')
pd.set_option('display.max_columns', 100)
#print(df.head())
#print(df.isnull().sum())
#print(df.shape)

#print('样本个数：{}'.format(df.shape[0]))
#print('男性个数：{}'.format(df[df.label=='male'].shape[0]))
#print('女性个数：{}'.format(df[df.label=='female'].shape[0]))
```

```python
df=pd.read_csv('voice.csv',index_col=0)
#print(train.head())
# 处理label字段
df['label']=df['label'].map(lambda x:1 if x=='male' else 0)
# 查看数据是否有空值
#print(train.isna().sum())

# 显示特征之间的相关系数
plt.figure(figsize=(10, 10))
plt.title('Pearson Correlation between Features',y=1.05,size=15)

# 计算特征之间的Pearson系数，即相似度
sns.heatmap(df.astype(float).corr(),linewidths=0.1,vmax=1.0, square=True,linecolor='white',annot=True)
plt.show()
```

```python
# 分离特征值与label
# ：-1 不要最后一列
X = df.iloc[:,:-1]
y = df.iloc[:,-1]

# 使用标签编码（x为数值类型，不需要进行标签编码）
gender_encoder = LabelEncoder()
y = gender_encoder.fit_transform(y)
print(y)

# 对x进行规范化
scaler = StandardScaler()
# fit：拟合 transform：赋值
X = scaler.fit_transform(X)

# 数据切分
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
svc = SVC()
svc.fit(X_train, y_train)
y_pred = svc.predict(X_test)
print('SVM 预测结果：', y_pred)
print('SVM 预测准确率：', accuracy_score(y_test,y_pred))
```
