```python
import datetime
import numpy as np
import pandas as pd
from collections import defaultdict
from datetime import datetime,timedelta
```

```python
# 数据加载
filename = './tianchi_fresh_comp_train_user.csv'
df = pd.read_csv(filename)
# 设置最大显示列数
pd.options.display.max_columns = 100
print(df.head())
```

```python
# 计算CVR(多少次点击产生一次购买)
count_all = 0
count_user = df['behavior_type'].value_counts()
count_all = count_user[1] + count_user[2] + count_user[3] + count_user[4]
count_4 = count_user[4]
cvr = count_4 / count_all
print('cvr={}%'.format(cvr*100))
```

```python
# 将time字段设置为pandas里面的datetime类型
df['time'] = pd.to_datetime(df['time'])
df.index = df['time']
# axis表示列，inplace表示修改原文(原地替换)
df.drop(['time'], axis=1, inplace=True)
print(df.head())
```

```python
# 时间规律统计
def show_count_day(df):
	count_day = defaultdict(int)
	# 从2014-11-18到2014-12-18
	str1 = '2014-11-17'
	# 字符串类型转换成时间类型
	temp_date = datetime.strptime(str1, '%Y-%m-%d')
	delta = timedelta(days=1)
	for i in range(30):
		temp_date = temp_date + delta
		# 时间类型转换为字符串类型
		temp_str = temp_date.strftime('%Y-%m-%d')
		print(temp_str)
		# df里面的index是时间
		count_day[temp_str] = count_day[temp_str] + df[temp_str].shape[0]
		print(count_day)
		
show_count_day(df)
```

```python
# 属于商品子集P的操作次数
df_p = pd.read_csv('./tianchi_fresh_comp_train_item.csv')
print(df_p.head())
# reset_index是将index还原，on表示按某列进行merge
print(df.shape)
df = pd.merge(df.reset_index(), df_p, on=['item_id']).set_index('time')
print(df.shape)
show_count_day(df)
```

```python
# 基于小时的统计分析，可以指定某个日期
def show_count_hour(date1):
	count_hour = {}
	# 设置初始值
	for i in range(24):
		time_str = date1 + ' %0.2d'%i
		count_hour[time_str] = [0, 0, 0, 0]
	
	for i in range(24):
		time_str = date1 + ' %0.2d'%i
		temp = df[time_str]['behavior_type'].value_counts()
		for j in range(len(temp)):
			count_hour[time_str][temp.index[j]-1] += temp[temp.index[j]]
		print(count_hour)
```

```python
show_count_hour('2014-12-10')
```
