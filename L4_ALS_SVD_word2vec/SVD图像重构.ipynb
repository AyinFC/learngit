# 导入np svd Image plt库

```python
import numpy as np
from scipy.linalg import svd
from PIL import Image
import matplotlib.pyplot as plt
import cv2
```

# 定义函数

```python
# 取前k个特征，对图像进行还原
def get_image_feature(s, k):
	# 对于S，只保留前K个特征值
	s_temp = np.zeros(s.shape[0])
	s_temp[0:k] = s[0:k]
	s = s_temp * np.identity(s.shape[0])
	# 用新的s_temp，以及p,q重构A
	temp = np.dot(p,s)
	temp = np.dot(temp,q)
	plt.imshow(temp, cmap=plt.cm.gray, interpolation='nearest')
	plt.show()
	print(matrix-temp)
```

# 加载图片

```python
image = Image.open('./255.jpg') 
A = np.array(image)
```

# 显示原图片

```python
plt.imshow(A, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

# 转化灰度图

```python
A.shape
matrix = A[:,:,1]
matrix.shape
plt.imshow(matrix, cmap=plt.cm.gray, interpolation='nearest')
plt.show()
```

# 对图像矩阵matrix进行奇异值分解，得到p,s,q

```python
p,s,q = svd(matrix, full_matrices=False)
```

# 取前k个特征，对图像进行还原

```python
get_image_feature(s, 5)
get_image_feature(s, 50)
get_image_feature(s, 500)
```
